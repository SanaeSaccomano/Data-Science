```python
df['packaging'].value_counts()
```

La variable "packaging" contient des informations sur le packaging des produits dans plusieurs langues avec plusieurs façons d'écrire le même emballage.
Il s'agit maintenant de créer une variable "packaging" qui regroupe plusieurs façons de décrire un emballage sous la même modalité.

```python
'''
Cette fonction détecte des mots issus d'un dictionnaire pour les changer, il remplace la valeur par 'other' si le mot n'est pas contenu dans le dictionnaire
Entrées :
entrees = df[variable]
dictionary = dictionnaire dont les clés sont les listes des valeurs acceptables pour la variable
defaut = remplaçant de la valeur si celle-ci ne se trouve pas dans les clés du dictionnaire

Renvoie :
result : liste des nouvelles valeurs    
'''
# Détection des mots issus d'un dictionnaire pour les changer, il remplace la valeur par 'other' si le mot n'est pas contenu dans le dictionnaire
df['packaging'] = my_desc.detect_words_other(df['packaging'], sp3.Dico_packaging_tags,
                                     'autre')
df['packaging'] = df['packaging'].str.capitalize()
```

```python
df['packaging'].value_counts()
```

Les emballages sont ensuite regroupés par matière dans une nouvelle variable "packaging_matiere".

```python
Dico_packaging_matiere = {
    'verre': 'verre',
    'plastique': 'plastique',
    'carton': 'carton',
    'conserve': 'metal',
    'papier': 'papier',
    'unknown': 'unknown',
}
```

```python
# Regroupement des packaging par matière
df['packaging_matiere'] = my_desc.detect_words_other(df['packaging'].str.lower(),
                                             Dico_packaging_matiere, 'autre')
df['packaging_matiere'] = df['packaging_matiere'].str.capitalize()
```

```python
df['packaging_matiere'].value_counts()
```

Il n'est pas nécessaire de garder les variables "packaging" et "packaging_matiere" dans la base de données. En effet, ce qui est intéressant ce sont les proportions de chacune de leurs modalités. Pour résumer celà, deux petites bases de données réunissant les proportions des modalités de chacune de ces variables sont créées.

```python
''' Cette fonction crée une petite base de données en comptant le nombre de modalités par variable
Entrées :
df : le dataframe
col : la variable d'origine
var1 : la variable de destination
var2 : la variable value_counts() de la variable col
reindexer : string si oui : opère une réindexation
liste_reindex : liste pour effectuer la reindexation

Renvoie :
data : une petite base de données correspondant aux modalités et leur nombre d'apparition au sein de la variable
'''

# La fonction mini_bdd permet de construire une base de données à partir d'un value_counts()
data_packaging = my_desc.mini_bdd(df,
                          'packaging',
                          "Emballage",
                          "Nombre",
                          reindexer='oui',
                          liste_reindex=[
                              'Unknown', 'Plastique', 'Sachet,plastique',
                              'Barquette,plastique', 'Autre', 'Bouteille',
                              'Carton', 'Brique,carton', 'Verre',
                              'Bocal,verre', 'Bouteille,verre', 'Conserve',
                              'Papier'
                          ])
data_packaging
```

```python
# La fonction mini_bdd permet de construire une base de données à partir d'un value_counts()
data_matiere = my_desc.mini_bdd(df, 'packaging_matiere', "Matière Emballage", "Nombre")
data_matiere
```
