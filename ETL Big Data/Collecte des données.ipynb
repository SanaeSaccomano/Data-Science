L'entreprise veut améliorer sa plateforme avec une nouvelle fonctionnalité de collaboration. Elle souhaite que ses utilisateurs puissent poster des avis et des photos de leur restaurant favori. 
Du point de vue de l'entreprise, c'est une stratégie pour mieux comprendre les avis postés par les usagers de la plateforme.

L'utilisateur peut :

- poster des avis sous forme de commentaires.
- poster des photos prises dans le restaurant.

La stratégie repose sur deux objectifs principaux:

- la détection des sujets d’insatisfaction présents dans les commentaires postés sur la plateforme
- la labellisation automatique des photos postées sur la plateforme (identifier les photos relatives à la nourriture, au décor dans le restaurant ou à l’extérieur du restaurant).

Le premier objectif repose sur des données textuelles (des lignes de caractères) et la construction d'un algorithme qui détecte les sujets d'insatisfactions dans les commentaires laissés par les usagers de la plateforme.
Le deuxième objectif repose sur les données visuelles (des images) et l'élaboration d'un algorithme de classification des photos postées par les usagers selon différents étiquettages (labels).

La preuve de concept de ce projet repose sur les données fournies par Yelp.
La première étape est donc de récolter ces données de façon efficace et optimale.
