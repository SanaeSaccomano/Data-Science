## Détermination des valeurs aberrantes

```python
# On calcule les bornes des valeurs aberrantes, leur proportion dans le dataframe complet
# et on crée un dataset sans les valeurs aberrantes
"""
    Cette fonction calcule les 1er et 3e quartiles pour trouver les valeurs aberrantes, 
    calcule les bornes inférieures et supérieures et renvoie un texte avec les résultats.

    Entrées :
    - df : le dataframe
    - colonne : la colonne d'intérêt
    - texte : si texte = 'oui', affiche le résultat sous forme de texte (par défaut = 'non')

    Renvoie :
    - la borne inférieure (q1 - 1.5 * IQR)
    - la borne supérieure (q3 + 1.5 * IQR)
    - un dataframe sans les valeurs abérrantes
    """

inf, sup, df_sva = my_desc.abbvalues(df, "Circonference_cm", 'oui')

inf, sup, df_sva = my_desc.abbvalues(df_sva, "Hauteur_m", 'oui')
```
